# Use of Power Transform : Box-Cox & YeoJohnson

```python
# power_transform supports the Box-Cox transform and the Yeo-Johnson transform
# Box-Cox requires input data to be strictly positive, while Yeo-Johnson supports both positive or negative data
```

```python
import numpy as np
from sklearn.preprocessing import power_transform
```

```python
df_train_sel
```

```python
df_train_sel_yeojohnson = power_transform(df_train_sel, method="yeo-johnson")
```

```python
df_train_sel_yeojohnson
```

```python
print((df_train_sel < 0).values.any())
```
